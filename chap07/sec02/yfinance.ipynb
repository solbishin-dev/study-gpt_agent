```python
import yfinance as yf

# Microsoft (MSFT)에 대한 Ticker(금융 상품 식별하는 고유 코드) 객체 생성
msft = yf.Ticker("MSFT")

# Ticker 객체에 대한 정보 출력 (.py에서 실행할 때는 print(msft.info)로 사용)
display(msft.info)
```

```python
hist = msft.history(period="2mo") # 2개월간의 주가 데이터를 가져옴
display(hist) # 데이터 출력
```
